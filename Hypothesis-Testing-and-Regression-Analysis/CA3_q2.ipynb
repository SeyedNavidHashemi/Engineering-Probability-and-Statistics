```python
import matplotlib.pyplot as plt
import numpy as np, random
def set_seed(seed):
  np.random.seed(seed)
  random.seed(seed)
set_seed(810109203)
```

#function for doing linier regression (least square)

```python
def fit_best_line_to_points(points):
  N = len(points)
  sigma_x = 0
  sigma_y = 0
  sigma_xy = 0
  sigma_x2 = 0
  for x, y in points:
    sigma_x += x
    sigma_y += y
    sigma_xy += (x * y)
    sigma_x2 += (x * x)

  m = ((N * sigma_xy) - (sigma_x * sigma_y)) / ((N * sigma_x2) - (sigma_x ** 2))
  b = (sigma_y - (m * sigma_x)) / N

  return m, b
```

#function for calculating Coefficient of Determination

```python
def calculate_coefficient_of_determination(points, m, b):
  SSR = 0
  y_bar = 0
  for x, y in points:
    y_bar += y
    SSR += ((y - ((m * x) + b)) ** 2)

  y_bar = y_bar / len(points)
  SST = 0
  for x, y in points:
    SST += ((y - y_bar) ** 2)

  R2 = 1 - (SSR / SST)
  return R2
```

#regression for 8 main points

```python
points_a = [(-2.3, -9.6), (-1.1, -4.9), (0.5, -4.1), (3.2, 2.7), (4.0, 5.9), (6.7, 10.8), (10.3, 18.9), (11.5, 20.5)]

m, b = fit_best_line_to_points(points_a)
R2_a = calculate_coefficient_of_determination(points_a, m, b)
x_points, y_points = zip(*points_a)

plt.scatter(x_points, y_points, color='red', label='Points')

x_line = np.linspace(min(x_points), max(x_points), 100)
y_line = (m * x_line) + b
plt.plot(x_line, y_line, linestyle='--', color='blue', label='fit line')

plt.text(-2, 15, f"R2 = {R2_a:.2f}", fontsize=12, color='black')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Points', color='b')

plt.legend()


plt.show()
```
